```python
import os
from langchain_community.llms.ollama import Ollama

# Load the locally hosted model
model_path = 'llama3.1'  # Ensure this path corresponds to the correct model name
llm = Ollama(model=model_path)

# Set the temperature parameter if needed
llm.temperature = 0

# Define a function to ask the LLM a question
def ask_llm_llama(prompt,patient_file):
    response = llm(prompt+patient_file)
    return response
```

```python
prompt = """
**Task:** Provide a tailored treatment plan for a breast cancer patient based on the provided medical file.

**Guidelines:**

* **Focus on:** Treatment options, considering the patient's specific condition, medical history, and preferences.
* **Prioritize:** Evidence-based approaches and personalized care.
* **Output format:** Clear and concise treatment plan, including recommended procedures, medications, and follow-up recommendations.

**Example:**

**Patient File:** [Insert patient file details here]

**Expected Output:**

* **Treatment Plan:**
    * Surgery: [Type of surgery, rationale]
    * Chemotherapy: [Regimens, dosage, schedule]
    * Radiation Therapy: [Area, dosage, schedule]
    * Targeted Therapy: [Agents, rationale]
    * Hormonal Therapy: [Type, rationale]
    * Follow-up: [Recommended schedule, tests]

**Additional Tips:**

* **Be specific:** Clearly define the patient's condition (e.g., stage, tumor type, hormone receptor status).
* **Provide context:** Include relevant medical history, comorbidities, and patient preferences.
* **Tailor the output:** Ensure the treatment plan aligns with the patient's individual needs and goals.
"""

```

```python
patient_file="""\n patient_file: A 62-year-old postmenopausal woman with ER-positive, HER2-negative metastatic breast cancer.
She has bone metastases but no visceral involvement.
She has been previously treated with an aromatase inhibitor (AI) and a CDK4/6 inhibitor.
"""
```

```python
import time
start_time = time.time()

print(ask_llm_llama(prompt,patient_file))

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time} seconds")
```
